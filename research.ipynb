```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/cleaned.csv')
```

```python
plt.figure(figsize=(12, 8))
df['Category'].value_counts().plot(kind='bar')
plt.title('Распределение новостей по категориям')
plt.show()
```

```python
df['Length'] = df['Content'].apply(lambda s: len(s.split()))
```

```python
plt.figure(figsize=(12, 8))
df.groupby('Category')['Length'].mean().plot(kind='bar')
plt.title('Средняя длина новости (в словах) по категориям')
plt.show()
```

```python
df.boxplot(column='Length', by='Category', figsize=(12, 8), rot=90)
plt.title('Длина новости (в словах) по категориям')
plt.show()
```
